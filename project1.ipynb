```python
3. Identify the predominant age group among employees.
```

```python
bins = [18, 30, 40, 50, 60, 100]
labels = ['18-30', '31-40', '41-50', '51-60', '60+']
df1['Age_group'] = pd.cut(df1['Age'], bins=bins, labels=labels, right=False)
age_group_distribution = df1['Age_group'].value_counts()
```

```python
age_group_distribution.plot(kind='bar', title='Predominant Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
4. Discover which team and position have the highest salary expenditure.
```

```python
team_salary_expenditure = df1.groupby('Team')['Salary'].sum()
position_salary_expenditure = df1.groupby('Position')['Salary'].sum()

max_team = team_salary_expenditure.idxmax()
```

```python
team_salary_expenditure.plot(kind='bar', title='Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.show()
```
